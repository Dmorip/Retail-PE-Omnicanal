```python
from pyspark.sql.functions import sum, avg, count, month, year,col
df_silver = spark.table("retail_peru_db.sales_silver")

df_gold = df_silver.groupBy("distrito", "tienda_nombre")\
    .agg(
        sum("monto_total").alias("total_ventas_pen"),
        avg("monto_total").alias("ticket_promedio"),
        count("id_transaccion").alias("num_transacciones")
    ).orderBy(col("total_ventas_pen").desc())

# 3. Guardar en Capa Gold
df_gold.write.format("delta").mode("overwrite").saveAsTable("retail_peru_db.sales_gold_performance")

display(df_gold)
```
